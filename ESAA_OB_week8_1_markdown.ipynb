### **1.3** 딥러닝이란

- 딥러닝: 인간의 신경망 원리를 모방한 심층 신경망 이론을 기반으로 고안된 머신 러닝 방법의 일종
- 인간의 뇌에 착안하여 컴퓨터에 뉴런(neuron), 시냅스(synapse) 개념 적용 (인간의 신경망 원리 모방) => 복잡하게 연결된 수많은 뉴런을 병렬 연산하여 기존에 컴퓨터가 수행하지 못했던 음성/영상 인식 등 처리를 가능케 함

##### **1.3.1 딥러닝 학습 과정**

1_7.PNG-

- 데이터 준비: 파이토치 제공 데이터셋, 캐글(Kaggle, AI 경진 대회 플랫폼) 등 활용
- 모델(모형) 정의: 신경망을 생성
- 모델(모형) 컴파일: 활성화 함수, 손실 함수, 옵티마이저 선택
  - 훈련 데이터셋 형태 연속형: 평균 제곱 오차(Mean Squared Error, MSE)
  - 훈련 데이터셋 형태 이진 분류: 크로스 엔트로피(cross, entropy)
- 모델(모형) 훈련: 한 번에 처리할 데이터양 지정 => 적당한 데이터양: 전체 훈련 데이터셋에서 일정한 묶음으로 나누어 처리할 수 있는 배치, 훈련 횟수(에포크) 선택이 중요
- 모델(모형) 예측: 검증 데이터셋을 생성한 모델(모형)에 적용하여 실제로 예측을 진행해 보는 단계

딥러닝 학습 과정 핵심 구성 요소
- 심층 신경망(deep neural network): 은닉층이 두 개 이상인 신경망
- 역전파: 가중치 값을 업데이트 => 오차를 각 가중치로 미분

##### **1.3.2 딥러닝 학습 알고리즘**

- 활용 분야 1. 지도 학습: 합성곱 신경망(Convolutional Neural Network)
  - 이미지 분류: 이미지를 알고리즘에 입력하면 그 이미지가 어떤 클래스에 속하는지 알려주어 이미지 데이터를 유사한 것끼리 분류
  - 이미지 인식: 사진을 분석하여 그 안에 있는 사물의 종류를 인식
  - 이미지 분할: 영상에서 사물이나 배경 등 객체 간 영역을 픽셀 단위로 구분하는 기술
  - 순환 신경망(Recurrent Neural Network, RNN): 시계열 데이터 분류
- 활용 분야 2. 비지도 학습
  - 워드 임베딩(word embedding): 자연어(사람의 언어)를 컴퓨터가 이해하고 효율적으로 처리할 수 있도록 자연어를 벡터로 표현
  - 군집: 아무런 정보가 없는 상태에서 데이터를 분류하는 방법 => 한 클러스터 안의 데이터는 매우 비슷하게 구성, 다른 클러스터의 데이터와 구분
- 활용 분야 3. 전이 학습(transfer learning): 사전에 학습이 완료된 모델(pre-trained model, 사전 학습 모델)을 가지고 우리가 원하는 학습에 미세 조정 기법을 이용하여 학습시키는 기법
  - 사전 학습 모델: 풀고자 하는 문제와 비슷하면서 많은 데이터로 이미 학습 되어 있는 모델

### **2.2** 파이토치 기초 문법

##### **2.2.1 텐서 다루기**


> **텐서 생성 및 변환**
- 텐서는 파이토치 가장 기본이 되는 데이터 구조
- 넘파이의 ndarray와 비슷, GPU 연산도 가능

````python
import torch
print(torch.tensor([[1,2],[3,4]]))   # 2차원 형태의 텐서 생성
print(torch.tensor([[1,2],[3,4]], device="cuda:0"))   # GPU에 텐서 생성
print(torch.tensor([[1,2],[3,4]], dtype=torch.float64))   # dtype을 이용하여 텐서 생성
````

````python
temp = torch.tensor([[1,2],[3,4]])
print(temp.numpy())   # 텐서를 ndarray로 변환

temp = torch.tensor([[1,2],[3,4]], device="cuda:0")
print(temp.to("cpu").numpy())   # GPU상의 텐서를 CPU의 텐서로 변환한 후 ndarray로 변환
````



> **텐서의 인덱스 조작**
- torch.FloatTensor: 32비트의 부동 소수점
- torch.DoubleTensor: 64비트의 부동 소수점
- torch.LongTensor: 64비트의 부호가 있는 정수



````python
temp = torch.FloatTensor([1,2,3,4,5,6,7])   # 파이토치로 1차원벡터생성
print(temp[0], temp[1], temp[-1])   # 인텍스로 접근
print('------------------------')
print(temp[2:5], temp[4:-1])   # 슬라이스로 접근
````



> **텐서 연산 및 차원 조작**
- 넘파이 ndarray처럼 다양한 수학 연산 가능, GPU 사용하면 더 빠른 연산 가능
- 텐서 간 타입이 다르면 연산이 불가능 (ex. FloatTensor - DoubleTensor)



````python
v = torch.tensor([1,2,3])   # 길이가 3인 벡터 생성
w = torch.tensor([3,4,6])
print(w-v)   # 길이가 같은 벡터 간 뺄셈 연산
````

- 텐서의 차원 변경 대표적인 방법: view (reshape과 유사)
- 텐서 결합 stack, cat => 다른 길이의 텐서 하나로 병합
- 텐서 차원 교환 t, transpose => 행렬의 전치, 차원의 순서 변경

````python
temp = torch.tensor([
    [1,2], [3,4]   # 2x2 행렬생성
])

print(temp.shape)
print('------------------------')
print(temp.view(4,1))             # 2x2 행렬을 4x1로 변형
print('------------------------')
print(temp.view(-1))              # 2x2 행렬을 1차원 벡터로 변형
print('------------------------')
print(temp.view(1, -1))           # 1x? 행렬을 temp 원소 개수(4) 유지한 채 1x4로 변형
print('------------------------')
print(temp.view(-1, 1))           # ?x1 행렬을 temp 원소 개수(4) 유지한 채 4x1로 변형
````

##### **2.2.3 모델 정의**
- 계층(layer): 모듈 또는 모듈을 구성하는 한 개의 계층 (합성곱층(convolutional layer), 선형계층(linear layer) ...)
- 모듈(module): 한 개 이상의 계층이 모여서 구성된 것, 모듈이 모여 새로운 모듈을 만들 수 있음
- 모델(model): 최종적으로 원하는 네트워크, 한 개의 모듈이 모델이 될 수 있음



> **단순 신경망을 정의하는 방법**

```
model = nn.Linear(in_features=1, out_features=1, bias=True)
```

> **nn.Module()을 상속하여 정의하는 방법**
- 파이토치에서 nn.Module 상속받는 모델: __ init __(), forward() 함수 포함
- __ init() __: 모델에서 사용될 모듈(nn.Linear, nn.Conv2d), 활성화 함수 등 정의
- forward(): 모델에서 실행되어야 하는 연산 정의






````python
from torch.nn import Module, Linear, Sigmoid

class MLP(Module):
    def __init__(self, inputs):
        super(MLP, self).__init__()
        self.layer = Linear(inputs, 1)   # 계층 정의
        self.activation = Sigmoid()   # 활성화 함수 정의

    def forward(self, X):
        X = self.layer(X)
        X = self.activation(X)
        return X
````



> **Sequential 신경망을 정의하는 방법**
- nn.Sequential: __ init __()에서 사용할 네트워크 모델 정의, forward() 함수에서 모델에서 실행되어야 할 계산 좀 더 가독성 뛰어나게 코드로 작성
- Seauential 객체는 그 안에 포함된 각 모듈을 순차적으로 실행
- nn.Sequential은 모델의 계층이 복잡할수록 효과가 뛰어남

````python
import torch.nn as nn
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.layer1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2)
        )

        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2)
        )

        self.layer3 = nn.Sequential(
            nn.Linear(in_features=30*5*5, out_features=10, bias=True),
            nn.ReLU(inplace=True)
        )

    def forward(self,x):
        x = self.layer1(x)
        x = self.layer2(x)
        x = x.view(x.shape[0], -1)
        x = self.layer3(x)
        return x

model = MLP()   # 모델에 대한 객체 생성

print("Printing Children\n------------------------------")
print(list(model.children()))
print("\n\nPrinting Modules\n------------------------------")
print(list(model.modules()))
````

- model.moules(): 모델의 네트워크에 대한 모든 노드를 반환
- model.children(): 같은 수준(level)의 하위 노드를 반환

> **함수로 신경망을 정의하는 방법**
- Sequential 이용하는 것과 동일하지만 함수로 선언할 경우 변수에 저장해 놓은 계층들을 재사용할 수 있음
- 복잡한 모델은 함수를 이용하는 것보다 nn.Module()을 상속받아 사용하는 것이 편리
- ReLU, Softmax, Sigmoid와 같은 활성화 함수는 모델 정의할 때 지정


````python
def MLP(in_features=1, hidden_features=20, out_features=1):
    hidden = nn.Linear(in_features=in_features, out_features=hidden_features,
                       bias = True)
    activation = nn.ReLU()
    output = nn.Linear(in_features=hidden_features, out_features=out_features,
                       bias=True)
    net = nn.Sequential(hidden, activation, output)
    return net
````

##### **2.2.4 모델의 파라미터 정의**
- 손실 함수(loss function): 학습하는 동안 출력과 실제 값(정답) 사이의 오차 측정
  - $wx+b$를 계산한 값과 실제 값인 $y$의 오차를 구해 모델의 정확성 측정
  - BCELoss: 이진 분류를 위해 사용
  - CrossEntropyLoss: 다중 클래스 분류를 위해 사용
  - MSELoss: 회귀 모델에서 사용
- 옵티마이저(optimizer): 데이터와 손실 함수를 바탕으로 모델의 업데이트 방법 결정
  - optimizer는 step() 메서드를 통해 전달받은 파라미터 업데이트
  - 모델의 파라미터별로 다른 기준(ex. 학습률)을 적용
  - torch.optim.Optimizer(params, defaults): 모든 옵티마이저의 기본이 되는 클래스
  - zero_grad ( ) 메서드: 옵티마이저에 사용된 파라미터들의 기울기(gradient)를 0으로 만듦
  - torch. optim .lr_scheduler: 에포크에 따라 학습률을 조절
  - 옵티마이저에 사용되는 종류
    - optim.Adadelta, optim.Adagrad , optim .Adam, optim.SparseAdam, optim.Adamax
    - optim.ASGD , optim.LBFGS
    - optim.RMSProp , optim.Rprop, optim.SGD
- 학습률 스케줄러(learning rate scheduler): 미리 지정한 횟수의 에포크를 지날 때마다 학습률을 감소(decay) 시켜 줌. 학습률 스케줄러를 이용하면 학습 초기에는 빠른 학습을 진행하다가 전역 최소점(global minimum) 근처에 다다르면 학습률을 줄여 최적점을 찾아갈 수 있도록 해줌
  - optim.lr_scheduler.LambdaLR: 람다(lambda) 함수를 이용하여 그 함수의 결괴를 학습률로 설정
  - optim.lr_scheduler.StepLR: 특정 단계(step) 마다 학습률을 감마(gamma) 비율만큼 감소
  - optim.lr_scheduler.MultiStepLR: StepLR과 비슷하지만 특정 단계가 아닌 지정된 에포크에만 감마 비율로 감소
  - optim.lr_scheduler.ExponentialLR: 에포크마다 이전 학습률에 감마만큼 곱함
  - optim.lr_scheduler.CosineAnnealingLR: 학습률을 코사인(cosine) 함수의 형태처럼 변화시킴 => 학습률이 커지기도 작아지기도 함
  - optim.lr_scheduler.ReduceLROnPlateau: 학습이 잘되고 있는지 아닌지에 따라 동적으로 학습률을 변화시킬 수 있음
- 지표(metrics): 훈련과 테스트 단계를 모니터링

** 전역 최소점과 최적점
- 손실 함수의 값을 최소화하는 가중치와 바이어스를 찾는 것이 학습 목표
- 전역 최소점(global minimum): 오차가 가장 작을 때의 값 (=최적점)
- 지역 최소점(local minimum): 전역 최소점을 찾아가는 과정에서 만나는 홀(hole), 옵티마이저가 지역 최소점에서 학습을 멈추면 최소값을 갖는 오차를 찾을 수 없는 문제 발생

##### **2.2.5 모델 훈련**
- 모델을 학습 => $y = wx + b$ 함수에서 $w$와 $b$의 적절한 값을 찾음
- $w$와 $b$에 임의의 값을 적용하여 오차가 줄어들어 전역 최소점에 이를 때까지 파라미터($w$, $b$)를 계속 수정
- 구체적 훈련 방법
  - optimizer.zero_grad() 메서드를 이용하여 기울기를 초기화
  - 기울기 값을 계산하기 위해 loss. backward() 메서드를 이용하는데, 새로운 기울기 값이 이전 기울기 값에 누적하여 계산 => 순환 신경망(Recurrent Neural Network, RNN) 모델을 구현할 때 효과적
  - 누적 계산이 필요하지 않는 모델에 대해서는 불필요 => optimizer.zero_grad() 메서드 호출: 미분 값이 누적되지 않게 초기화

- *loss.backward() 메서드를 이용하여 기울기를 자동 계산*
- loss.backward()는 배치가 반복될 때마다 오차가 중첩적으로 쌓여 매번 zero_grad() 사용하여 미분 값을 0으로 초기화
